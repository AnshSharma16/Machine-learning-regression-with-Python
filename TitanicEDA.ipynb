```python
from google.colab import files
uploaded = files.upload()

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop('Cabin', axis=1, inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
plt.scatter(df['Age'], df['Fare'])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
df['Age'].plot.hist(color='orange',edgecolor='Black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
survivors= df[df['Survived']==1]
sns.histplot(data= survivors, x= survivors['Age'], bins= 20, hue=survivors['Sex'],  palette= ['Blue','Magenta'])
plt.title("AGE OF SURVIVORS")
plt.xlabel('AGE')
plt.ylabel('COUNT')
plt.show()
```

```python
casualties= df[df['Survived']==0]
sns.histplot(data=casualties, x= casualties['Age'], bins= 20, hue=casualties['Sex'],  palette= ['Blue','Magenta'])
plt.title("AGE OF CASUALTIES")
plt.xlabel('AGE')
plt.ylabel('COUNT')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df, palette=['Skyblue','Magenta'])
plt.title('Sex vs Survived')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df, palette=['Red','Blue','Green'])
plt.title('Pclass vs Survived')
plt.show()
```

```python
sns.barplot(x='Embarked', y='Survived', data=df, palette=['Brown','Black','Grey'])
plt.title('Embarked vs Survived')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df, palette=['yellow', 'red'])
plt.title('Age vs Survived')
plt.show()
```

```python
sns.kdeplot(x='Age', hue='Survived', data=df, multiple='stack')
plt.title('Age vs Survived')
plt.show()
```

```python
sns.kdeplot(x='Fare', hue='Survived', data=df, multiple='stack')
plt.title('Fare vs Survived')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
df.head()
```

```python
df1=df.drop(['PassengerId','Name','Ticket'], axis=1)
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
